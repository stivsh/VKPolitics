```python
def load_posts_wall(owner_id):
    api = vk.API(access_token=token)
    items = []
    error = 1
    while error > 0:
        try:
            response = api('wall.get', owner_id=owner_id,filter='owner',count=100)
            if 'items' in response:
                items += response['items']
            error = 0
        except Exception,e:
            error += 1
            if error > 2:
                print str(e)
                return items
            time.sleep(2)
            
    return items

def get_posts_from_walls(group_name,ids):
    file_name = "vkdata/%s_walls_posts.pkl" % group_name
    if os.path.isfile(file_name):
        return pickle.load(open(file_name,"r"))
    else:
        items = {}
        for inx,id in enumerate(ids):
            items[id] = load_posts_from_wall(id)
            if inx % (len(ids)/10) == 0:
                print inx/(len(ids)/10) * 10, "% loaded"
        pickle.dump(items,open(file_name,"w"))
        print "%s posts loaded" % file_name
        return items
```

```python
def get_group_info(id):
    def load_group_info(id):
        api = vk.API(access_token=token)
        response = []
        error = 1
        while error > 0:
            try:
                response = api('groups.getById', group_ids=[id])
                error = 0
            except Exception,e:
                error += 1
                if error > 2:
                    print str(e)
                    return response
                time.sleep(2)
        return response
    
    if type(id)==type(''):
        id = int(id)
    if id <0: id*=-1
    inf = []
    file_name = "vkdata/groups_info.pkl"
    if not os.path.isfile(file_name):
        pickle.dump({},open(file_name,"w"))
    groups_info = pickle.load(open(file_name,"r"))
    if id in groups_info:
        inf = groups_info[id]
    else:
        inf = load_group_info(id)[0]
        groups_info[id] = inf
        pickle.dump(groups_info,open(file_name,"w"))
    
    return inf
        
        
    
```

```python
def get_top_sourses(group_name,ids):
    posts_of_users = get_posts_from_walls(group_name,ids)
    all_reposts_owners = [ post['copy_history'][0]['owner_id'] for post in sum([ val for key,val in posts_of_users.iteritems() ]) if 'copy_history' in post ]
    top_owners = {}
    for own in all_reposts_owners:
        top_owners[own] = top_owners.get(own,0)+1
    top = sorted(list(top_owners.iteritems()),key = lambda x:-x[1])[:10]
    return [ (x[0]['name'],x[1]) for x in [ (get_group_info(t[0]),t[1])for t in top ] ]
    
```

```python
rrp_owners = get_top_sourses('realproblemsrussia',rrp_ids)
nod_owners = get_top_sourses('rusnod',nod_ids)
ros_owners = get_top_sourses('rospil',ros_ids)
star_owners = get_top_sourses('nstarikov_club',star_ids)
```

```python
def plot_bar(ax,val,labels,title,color='b',rotation='horizontal'):
    ax.bar(range(len(val)), val, align='center',color=color, alpha=0.4)
    ax.xaxis.set_ticklabels(labels,rotation=rotation)
    ax.xaxis.set_ticks(range(len(val)))
    ax.set_title(title)
    ax.set_ylabel('%')
    
def get_plots(count,figsize=(11,8),layout=None):
    if layout is None:
        layout=(count,1)
    fig = plt.figure(figsize=figsize)
    return [fig]+[ fig.add_subplot(layout[0],layout[1],i+1) for i in range(count) ] 
    
fig,p1,p2,p3,p4 = get_plots(4,figsize=(15,15),layout=(2,2))
for p in [ (p1,"RialRusProb","b",rrp_owners),(p2,"Rospil","g",ros_owners),(p3,"NOD","r",nod_owners),(p4,"Starikov","m",star_owners) ]:
    plot_bar(p[0],[ z[1] for z in p[3] ],[ z[0] for z in p[3] ],p[1],p[2],rotation='vertical')
fig.tight_layout()
plt.show()    
```
